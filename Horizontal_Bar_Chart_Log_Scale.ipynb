```python
from matplotlib import *
from matplotlib import pyplot as plt
%matplotlib notebook
import pandas as pd
import numpy as np
np.random.seed(100)
pltparams = {
    'axes.labelsize': 20,
    'font.size': 18,
    'legend.fontsize': 15,
    'xtick.labelsize': 18,
    'ytick.labelsize': 18,
    'text.usetex': False,
    'figure.figsize': [10, 6]
}

importances = {'Feature_{}'.format(i): np.random.rand()*1000 for i in range(10)}


rcParams.update(pltparams)
plt.gca().set_frame_on(False)
plt.grid()
plt.xscale('log')

plt.barh(range(len(importances.keys())), importances.values(), color='g', align='center',
            alpha=0.4)

plt.yticks(range(len(importances.keys())), importances.keys(), size='small')

plt.grid(True)
plt.xlabel('Feature Importance')
plt.tight_layout()
plt.savefig('Horizontal_Bar_Chart_Log_Scale.pdf')
plt.savefig('Horizontal_Bar_Chart_Log_Scale.png')

plt.show()
```
