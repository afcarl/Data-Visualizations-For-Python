```python
from matplotlib import *
from matplotlib import pyplot as plt
%matplotlib notebook
import pandas as pd
import numpy as np
np.random.seed(100)
pltparams = {
    'axes.labelsize': 20,
    'font.size': 18,
    'legend.fontsize': 15,
    'xtick.labelsize': 18,
    'ytick.labelsize': 18,
    'text.usetex': False,
    'figure.figsize': [10, 6]
}

importances = {'Feature_{}'.format(i): np.random.rand()*1000 for i in range(10)}

importances = pd.DataFrame(importances.values(), index= importances.keys())
importances = importances.sort_values(importances.columns[0])
print(importances)

rcParams.update(pltparams)
plt.gca().set_frame_on(False)
plt.grid()
plt.xscale('log')

importances.plot.barh(ax = plt.gca(),color='g',
            alpha=0.4, legend=False)

plt.grid()
plt.xlabel('Feature Importance')
plt.tight_layout()
plt.savefig('Horizontal_Bar_Chart_Log_Scale.pdf')
plt.savefig('Horizontal_Bar_Chart_Log_Scale.png')

plt.show()
```
