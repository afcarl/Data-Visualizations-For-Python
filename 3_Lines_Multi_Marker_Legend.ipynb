```python
from matplotlib import *
from matplotlib import pyplot as plt
%matplotlib notebook

import numpy as np

pltparams = {
    'axes.labelsize': 15,
    'font.size': 8,
    'legend.fontsize': 20,
    'xtick.labelsize': 15,
    'ytick.labelsize': 15,
    'text.usetex': False,
    'grid.linestyle': 'dotted',
    'figure.figsize': [8, 5]
}

x = np.exp(-2*np.arange(0,1,.1))

y = np.exp(-4*np.arange(0,1,.1))
z = np.exp(-15*np.arange(0,1,.1))


rcParams.update(pltparams)

plt.figure()
ax = plt.gca()
ax.set_frame_on(False)

plt.grid()


plt.plot(range(len(x)), x, linestyle='dotted', marker='x', label=r'$S_x$')
plt.plot(range(len(y)), y, linestyle='dashed', marker='^', label=r'$S_y$')
plt.plot(range(len(z)), z, linestyle='dashed', marker='o', label=r'$S_z$')


ax.set_xlabel(r'$\tau$')
ax.set_ylabel('Probability')
plt.ylim((-0.1,1.1))
plt.xlim((-0.1,9.1))

plt.legend()
plt.tight_layout()
plt.savefig('3_Lines_Multi_Marker_Legend.pdf')
plt.savefig('3_Lines_Multi_Marker_Legend.png')

plt.show()
```
