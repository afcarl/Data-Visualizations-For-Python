```python
"""
Heavily based on this great tutorial:
https://github.com/jbmouret/matplotlib_for_papers
"""

from matplotlib import *
from matplotlib import pyplot as plt
%matplotlib notebook

import numpy as np
pltparams = {
    'axes.labelsize': 8,
    'font.size': 8,
    'legend.fontsize': 15,
    'xtick.labelsize': 10,
    'ytick.labelsize': 10,
    'text.usetex': False,
    'grid.linestyle': 'dotted',
    'figure.figsize': [10, 4]
}

rcParams.update(pltparams)

plt.figure()



ax = plt.subplot(211)
ax.set_frame_on(False)
ax.grid()
data = np.sin([x for x in np.arange(0,50,0.01)])
plt.plot(data, c='g', alpha=.4)

#plt.setp( ax.get_xticklabels(), visible=False) uncomment if you want to hide xticks labels
ax.xaxis.get_offset_text().set_visible(False)

ax = plt.subplot(212)
ax.set_frame_on(False)


data = np.cos([x for x in np.arange(0,100,0.01)])
plt.plot(data, c='r', alpha=.4)
#plt.setp( ax.get_xticklabels(), visible=False) uncomment if you want to hide xticks labels
ax.xaxis.get_offset_text().set_visible(False)
ax.grid()

plt.tight_layout()
plt.savefig('2_subplots_vertically_stacked.pdf')
plt.show()

```
