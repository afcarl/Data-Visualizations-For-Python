```python
from matplotlib import *
from matplotlib import pyplot as plt
%matplotlib notebook

import numpy as np
pltparams = {
    'axes.labelsize': 10,
    'font.size': 15,
    'legend.fontsize': 15,
    'xtick.labelsize': 10,
    'ytick.labelsize': 10,
    'text.usetex': False,
    'grid.linestyle': 'dotted',
    'figure.figsize': [10, 4]
}

X = np.random.rand(500)

rcParams.update(pltparams)
plt.gca().set_frame_on(False)
plt.grid()


plt.scatter(np.where(X<.5), X[X<.5],
            marker='o', color='g', s=40,
            alpha=0.4, label='ok')

plt.scatter(np.where(X>=.5), X[X>=.5],
            marker='x', color='r', s=40, alpha=.4,
            label='error')

plt.axhline(0.5, c='b', alpha=.4)

plt.xlabel('Id')
plt.ylabel('Probability')
plt.tight_layout()
plt.savefig('2_Categories_Scatter_Plot.pdf')
plt.show()
```
